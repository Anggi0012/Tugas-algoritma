```python
import numpy as np
import matplotlib.pyplot as plt

def garis_dua_titik_28(x1_28, y1_28, x2_28, y2_28):
    m_28 = (y2_28 - y1_28) / (x2_28 - x1_28)
    c_28 = y1_28 - m_28 * x1_28
    return m_28, c_28

def garis_gradien_titik_28(m_28, x_28, y_28):
    c_28 = y_28 - m_28 * x_28
    return m_28, c_28

def garis_persamaan_umum_28(a_28, b_28, e_28):
    if b_28 == 0:
        raise ValueError("Nilai b tidak boleh nol pada persamaan ax + by = e.")
    m_28 = -a_28 / b_28
    return m_28, e_28 / b_28

def input_dua_titik_28():
    x1_28 = float(input("Nilai x dari titik pertama adalah "))
    y1_28 = float(input("Nilai y dari titik pertama adalah "))
    x2_28 = float(input("Nilai x dari titik kedua adalah "))
    y2_28 = float(input("Nilai y dari titik kedua adalah "))
    return garis_dua_titik_28(x1_28, y1_28, x2_28, y2_28)

def input_gradien_titik_28():
    m_28 = float(input("Nilai gradien adalah "))
    x_28 = float(input("Nilai x dari titik yang dilewati adalah "))
    y_28 = float(input("Nilai y dari titik yang dilewati adalah "))
    return garis_gradien_titik_28(m_28, x_28, y_28)

def input_persamaan_umum_28():
    a_28 = float(input("Nilai a pada persamaan ax + by = e adalah "))
    b_28 = float(input("Nilai b pada persamaan ax + by = e adalah "))
    e_28 = float(input("Nilai e pada persamaan ax + by = e adalah "))
    return garis_persamaan_umum_28(a_28, b_28, e_28)

def input_garis_28():
    print("Pilih jenis input garis:")
    print("1. Garis melewati dua titik")
    print("2. Garis diketahui gradien dan satu titik yang dilewati")
    print("3. Persamaan garis umum ax + by = e")
    pilihan_28 = int(input("Masukkan pilihan (1, 2, atau 3): "))

    if pilihan_28 == 1:
        return input_dua_titik_28()
    elif pilihan_28 == 2:
        return input_gradien_titik_28()
    elif pilihan_28 == 3:
        return input_persamaan_umum_28()
    else:
        raise ValueError("Pilihan tidak valid.")

print("Input garis pertama:")
m1_28, c1_28 = input_garis_28()

print("Input garis kedua:")
m2_28, c2_28 = input_garis_28()

print("Persamaan garis pertama adalah y={}x+{}".format(m1_28, c1_28))
print("Persamaan garis kedua adalah y={}x+{}".format(m2_28, c2_28))

if m1_28 == m2_28:
    if c1_28 == c2_28:
        print("Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi.")
    else:
        print("Kedua garis sejajar sehingga tidak terdapat solusi.")
else:
    tp_x_28 = (c2_28 - c1_28) / (m1_28 - m2_28)
    tp_y_28 = m1_28 * tp_x_28 + c1_28
    print("Titik potong kedua garis adalah ({},{})".format(tp_x_28, tp_y_28))

x_28 = np.linspace(-10, 10, 100)
plt.plot(x_28, m1_28 * x_28 + c1_28, label="Garis 1")
plt.plot(x_28, m2_28 * x_28 + c2_28, label="Garis 2")
plt.axvline(x=0, color='grey', lw=1)
plt.axhline(y=0, color='grey', lw=1)
if m1_28 != m2_28:
    plt.plot(tp_x_28, tp_y_28, 'ro')  # titik potong
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
