```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_from_roots_28(root1_28, root2_28):
    # Menyusun persamaan kuadrat dari akar-akarnya
    a_28 = 1
    b_28 = -(root1_28 + root2_28)
    c_28 = root1_28 * root2_28
    return a_28, b_28, c_28

def quadratic_function_28(a_28, b_28, c_28, x_28):
    # Menghitung nilai fungsi kuadrat
    return a_28 * x_28**2 + b_28 * x_28**2 + c_28

def vertex_and_axis_28(a_28, b_28, c_28):
    # Menentukan titik puncak dan sumbu simetri
    x_vertex_28 = -b_28 / (2*a_28)
    y_vertex_28 = a_28 * x_vertex_28**2 + b_28 * x_vertex_28 + c_28
    return x_vertex_28, y_vertex_28, -b_28 / (2*a_28)

def discriminant_28(a_28, b_28, c_28):
    # Menghitung diskriminan
    return b_28**2 - 4*a_28*c_28

def solve_quadratic_28(a_28, b_28, c_28):
    disc_28 = discriminant_28(a_28, b_28, c_28)
    if disc_28 > 0:
        root1_28 = (-b_28 + np.sqrt(disc_28)) / (2*a_28)
        root2_28 = (-b_28 - np.sqrt(disc_28)) / (2*a_28)
        return root1_28, root2_28
    elif disc_28 == 0:
        root_28 = -b_28 / (2*a_28)
        return root_28,
    else:
        real_part_28 = -b_28 / (2*a_28)
        imaginary_part_28 = np.sqrt(-disc_28) / (2*a_28)
        return real_part_28 + imaginary_part_28*1j, real_part_28 - imaginary_part_28*1j

def plot_quadratic_28(a_28, b_28, c_28):
    x_28 = np.linspace(-10, 10, 400)
    y_28 = quadratic_function_28(a_28, b_28, c_28, x_28)

    plt.figure(figsize=(8, 6))
    plt.plot(x_28, y_28, label=f"{a_28}x^2 + {b_28}x + {c_28}")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

    # Titik puncak dan sumbu simetri
    x_vertex_28, y_vertex_28, axis_of_symmetry_28 = vertex_and_axis_28(a_28, b_28, c_28)
    plt.plot(x_vertex_28, y_vertex_28, 'ro')  # Titik puncak
    plt.axvline(x=axis_of_symmetry_28, color='r', linestyle='--', label=f'Sumbu Simetri: x = {axis_of_symmetry_28}')

    plt.legend()
    plt.title("Grafik Persamaan Kuadrat")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

def main():
    print("Program Analisis dan Grafik Persamaan Kuadrat")
    root1_28 = float(input("Masukkan akar pertama: "))
    root2_28 = float(input("Masukkan akar kedua: "))

    a_28, b_28, c_28 = quadratic_from_roots_28(root1_28, root2_28)
    print(f"Persamaan kuadrat: {a_28}x^2 + {b_28}x + {c_28}")

    # Titik puncak dan sumbu simetri
    x_vertex_28, y_vertex_28, axis_of_symmetry_28 = vertex_and_axis_28(a_28, b_28, c_28)
    print(f"Titik Puncak: ({x_vertex_28}, {y_vertex_28})")
    print(f"Sumbu Simetri: x = {axis_of_symmetry_28}")

    # Posisi titik puncak relatif terhadap sumbu y
    if x_vertex_28 < 0:
        print("Titik puncak berada di kiri sumbu y")
    elif x_vertex_28 > 0:
        print("Titik puncak berada di kanan sumbu y")
    else:
        print("Titik puncak berada pada sumbu y")

    # Kondisi parabola memotong sumbu y
    if c_28 > 0:
        print("Parabola memotong sumbu y di atas (positif)")
    elif c_28 < 0:
        print("Parabola memotong sumbu y di bawah (negatif)")
    else:
        print("Parabola memotong sumbu y di pusat (0, 0)")

    # Macam-macam akar persamaan kuadrat
    disc_28 = discriminant_28(a_28, b_28, c_28)
    print(f"Diskriminan: {disc_28}")
    roots_28 = solve_quadratic_28(a_28, b_28, c_28)
    if disc_28 > 0:
        print(f"Akar-akarnya adalah real dan berbeda: {roots_28}")
    elif disc_28 == 0:
        print(f"Ada satu akar real: {roots_28}")
    else:
        print(f"Akar-akarnya adalah kompleks: {roots_28}")

    # Grafik persamaan kuadrat
    plot_quadratic_28(a_28, b_28, c_28)

if __name__ == "__main__":
    main()
```
